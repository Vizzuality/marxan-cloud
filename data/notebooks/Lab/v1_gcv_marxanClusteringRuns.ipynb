```r
library(vegan)
```

```r
mat = matrix(data = sample(c(0,1),size = 100000000, replace = T), nrow = 10000, ncol = 10000, 
       dimnames =list (paste0("run_", 1:10000), paste0("pu_", 1:10000)))
```

```r
mat_dis = vegdist(mat, method = "jaccard")
```

```r
hc = hclust(mat_dis)
plot(hc)
rect.hclust(hc, k=5, border = "red")
```

```r
str(hc)
```

```r
cluster = cutree(hc, 5)
head(cluster)
```
